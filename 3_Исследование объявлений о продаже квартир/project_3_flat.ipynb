### 2.1 Определите и изучите пропущенные значения:

```python
#Заменили значения NaN в столбце `balcony`, `days_exposition`, на 0. 
#Так как число балконо нету и наверно обьявление весит меньше суток.
data['balcony'] = data['balcony'].fillna(0)
data['days_exposition'] = data['days_exposition'].fillna(0)
# Пустые строки в столбце is_apartment образовались из-за того, что продавец просто не указал что это не квартира 
# или эти аппартаменты появились совсем недавно
data['is_apartment'] = data['is_apartment'].fillna(False)
# В locality_name ничего не будем менять, так как причина не известна

# В ceiling_height видно мега высокие потолки, доходящие до 100 метов, их причина тоже не известна, 
# только если это чисто один небоскреб

# В floors_total ничего обычного не видим, поэтому тоже оставляем как есть, а то если гадать сколько этажей в доме
# можно сбить полный анализ
```

### 2.2 Приведите данные к нужным типам:
- Поясните, в каких столбцах нужно изменить тип данных и почему.

```python
#Привели в порядок дату и время
#UPD: Время впринципе можно не указывать, его, грубо говоря, там нет. Но я оставлю
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format=('%Y-%m-%dT%H:%M:%S'))

#Заменил тип для столбца `balcony`, `days_exposition`, `last_price` для экономия мест 
data['balcony'] = data['balcony'].astype('int32')
data['days_exposition'] = data['days_exposition'].astype('int32')
data['last_price'] = data['last_price'].astype('int32')
```

## Шаг 3. Посчитайте и добавьте в таблицу
- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры; варианты — первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.


```python
#Квадратный метр = Цена/площадь 
#Меняем на тип `int` для удобства чисел
data['price_meter'] = data['last_price'] / data['total_area']
data['price_meter'] = data['price_meter'].astype('int')
#Добавляем столбцы в таблицу со значениями: день недели, месяца и года публикации.
data['w_exposition'] = data['first_day_exposition'].dt.weekday
data['m_exposition'] = data['first_day_exposition'].dt.month
data['y_exposition'] = data['first_day_exposition'].dt.year
#Категоризация данных по этажам
def floor_cat(row):
    floors_total = row['floors_total']
    floor = row['floor']
    
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    elif 1 < floor < floors_total:
        return 'другой'
data['floor_cat'] = data.apply(floor_cat, axis=1)
#Добавляем два столбца: 1 - жил.площ/общ.площ, 2 - кухн.площ/общ.площ
data['total_living'] = data['living_area'] / data['total_area']
data['total_kitchen'] = data['kitchen_area']/data['total_area']
```

```python
#Зависимость цена от площади
price_meter_total_area = data.pivot_table(index='total_area', values='price_meter', aggfunc=['median'])
price_meter_total_area.plot(y='median', style='o', alpha=0.5)
data['total_area'].corr(data['price_meter'])
```
